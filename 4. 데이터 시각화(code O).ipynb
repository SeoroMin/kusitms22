## 실습

```python
# 이미지 크기 변경
plt.figure(figsize=(10,5))

# plot 그리기(x값, y값, 선형태, 색깔)
plt.plot(["Seoul", " Paris", "Seattle"], [10, 35, 20], '-', color="orange")
plt.plot(["Seoul", " Paris", "Seattle"], [20, 45, 50], '--', color="blue")
plt.plot(["Seoul", " Paris", "Seattle"], [40, 20, 10], '--', color="yellow")

# x축 이름 지정
plt.xlabel('city')

# y축 이름 지정
plt.ylabel("values")

# 제목 지정
plt.title("Example")

# 범례 추가 <- 범례를 어디다 둘지
# upper, center, lower / left, center, right
plt.legend(['a','b','c'], loc='upper left')

# 축값 변경 -> 축값의 범주로 들어감.
plt.ylim(0,60)

# grid 표사
plt.grid(True)
```

## subplot() - 한 화면에 여러 그래프 동시에 그리기

```python
# figsize 지정
fig = plt.figure(figsize=(5,5))
#여러개 그리려면 크기를 일단 변수 하나에 저장을 해 줘야함.

# ax1
ax1 = fig.add_subplot(2,2,1)
#몇 행 몇열을 그릴것이냐 이 경우 2x2임. <-- (2,2,1)은 2행 2열의 첫번째에다가 넣겠다는 의미.

# ax2
ax2 = fig.add_subplot(2,2,2)
#1행 2열에 ax2를 그리기

# ax3
ax3 = fig.add_subplot(2,2,3)

# ax4
ax4 = fig.add_subplot(2,2,4)

# histogram 그리기
ax1.hist(random.sample(range(1,100),50), bins = 10, color = 'y', alpha = 1)

#alpha는 투명도임. bins는 구간을 어떤 단위로 나눌것이냐. alpha는 겹친값을 확인해주려 이용함


# bar 그리기
ax2.bar(x=sr.index, height=sr2.values)

# scatter 그리기
ax3.scatter(random.sample(range(1,100), 50),
           random.sample(range(1,100),50))
#변수가 여러개 많으면 변수 두개를 골라서 산점도를 그려보면 값이 얼마나 퍼져있는구나를 알 수 있음
#데이터 컬럼을 선택하는데 있어서 많이 씀

# ax4 위치에 y=x 그래프를 그려보기
ax4.plot([x for x in range(10)])

#리스트 생성해줘서 바로 값을 주면된다.

plt.subplots_adjust(hspace=0.5, wspace = 0.3)
# hspace = height space 우아래, wspace = width space 양옆
```

# seaborn 시각화

```python
# tips 데이터 불러오기
tips = sns.load_dataset("tips")
```

```python
tips
```

## 1. relplot(scatter, line을 그려주는 함수)
## default : scatter

## (1) scatter plot(산점도)

```python
# x : x축 column
# y : y축 column
# data : 사용할 data frame
sns.relplot(x="total_bill", y="tip",data=tips)
```

```python
# hue(명목형) : 해당 변수를 기준으로 색깔 다르게 표시
sns.relplot(x="total_bill", y="tip", hue="sex",data=tips)
```

```python
# hue(실수형) : 해당 변수를 기준으로 색깔 그라데이션으로 표시
sns.relplot(x="total_bill", y="tip", hue="size",data=tips)
```

```python
# style: 해당 point의 marker를 다르게 표시
sns.relplot(x="total_bill", y="tip", hue="sex",style = "smoker", data=tips)
```

```python
# size: 점의 크기 조정
sns.relplot(x="total_bill", y="tip", hue="sex",style = "smoker",size='size', data=tips)
```

#### 점이 여러개 찍혀있을때는 점이 가려지는 경우가 많은데 그때 alpha를 써서 값을 조금 만져준다

```python
# alpha: 점의 투명도 조정(0~1)
sns.relplot(x="total_bill", y="tip", hue="sex",style = "smoker",size='size',alpha = 0.7, data=tips)
```

```python
# col : 선택된 변수에 따라서 subplot을 그려준다
sns.relplot(x="total_bill", y="tip", hue="sex", col="smoker", data=tips)
```

## (2) line plot

```python
# x : x축 column
# y : y축 column
# data : 사용할 data frame
# hue : 집단별 구분
# style : 해당 point의 marker를 다르게 표시
sns.relplot(x='tip', y='total_bill', hue='sex',style='sex', kind='line', data=tips)
```

## 2. catplot() : 변수가 명목형일 때 사용

## (1) Strip plot

- 순서를 바꿔서 봐주고싶다 --> order옵션을 사용
- 특정값에 몰려있을때 분포를 보기위해서 jitter라는 옵션을 사용한다

```python
# kind : 'strip'
# order : 명목형 범주 지정
# jitter : 점이 겹쳐있을 때 살짝 떨어져서 보이게 만드는 기능
sns.catplot(x='smoker', y='tip', jitter=False, order=['No', 'Yes'], data=tips)
```

## (2) boxplot

```python
# kind : 'box'
sns.catplot(x='smoker', y='tip', hue='sex', kind='box', data=tips)
```

## (3) barplot

```python
# kind : 'bar'
# dodge : 집단 막대 그래프를 나누어서 그림(default:True)
sns.catplot(x="smoker", y="tip", hue="sex", kind="bar", ci=None, dodge = True, data=tips)
```

- bar 위에 박혀있는 라인은 신뢰구간을 나타내는것임
- ci = None을 통해서 신뢰구간 없앰
- dodge는 두 그래프를 합쳐준다
- 약간 stack같은 느낌

## 3. countplot() : 명목형 변수의 개수를 막대그래프로 시각화
